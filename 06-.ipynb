```python
import pickle
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('dblp_filtered.pickle', 'rb') as f:
    dblp_dic = pickle.load(f)
```

```python
with open('faculties_by_interests.json', 'r') as f:
    faculties_dic = json.load(f)
```

```python
dic = {}
for univ, fac in faculties_dic.items():
    dic[univ] = len(fac)
pd.Series(dic).sort_values().plot.barh(figsize=(6, 12))
```
