```python
with open('../CSrankings/dblp.xml', 'r', encoding="ISO-8859-1") as f:
    test = xmltodict.parse(f.read())
```

```python
test['dblp']['phdthesis'][0]
```

```python
pa_dic = {}
for k, li in test['dblp'].items():
    pa_dic[k] = []
    for dic in li:
        if 'author' in dic.keys():
            author = dic['author']
            if 'Pieter Abbeel' in author:
                pa_dic[k].append(dic)
        
```
